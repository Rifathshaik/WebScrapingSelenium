```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#creating driver object

driver=webdriver.Chrome(r"C:/chromedriver.exe")
```

```python
#getting the driver
driver.get("https://www.flipkart.in/")
```

```python
#Searching for smart phone
Search=driver.find_element_by_xpath("//input[@title='Search for products, brands and more']")
Search.send_keys("smart phone")
```

```python
#clicking search button through code.
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
#fetching urls of the products
product_urls=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
```

```python
#creating empty list to add urls of each product
prod_url=[]
```

```python
#adding urls to the list
for i in range(len(product_urls)):
            prod_url.append(product_urls[i].get_attribute("href"))
```

```python
len(prod_url)
```

```python
#creating empty list to store values
brand=[]
smartphone=[]
color=[]
RAM=[]
ROM=[]
pri_cam=[]
sec_cam=[]
disp_size=[]
disp_res=[]
prcsr=[]
prcsr_cores=[]
batry=[]
price=[]
url=[]

            
```

```python
for i in range(len(prod_url)):
    #getting driver for each products page
    driver.get(prod_url[i])
    url.append(prod_url[i])
    name=driver.find_element_by_xpath('//h1[@class="yhB1nd"]')
    brand.append(name.text)
    #clicking read more in the page _2KpZ6l _1FH0tX Read More
    readmore=driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]')
    readmore.click()
    try:
        phnname=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        smartphone.append(phnname.text)
    except NoSuchElementException:
        smartphone.append('-----')
    
    try:
        colors=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        color.append(colors.text)
    except NoSuchElementException:
        color.append('-----')
    try:
        dispSize=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        disp_size.append(dispSize.text)
    except NoSuchElementException:
        disp_size.append('-----')
    try:
        dispres=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[2]/td[2]/ul/li')
        disp_res.append(dispres.text)
    except NoSuchElementException:
         disp_res.append('-----')
    try:
        processor=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[3]/table/tbody/tr[2]/td[2]/ul/li')
        prcsr.append(processor.text)
    except NoSuchElementException:
        prcsr.append('-----')
    try:
        procesor_core=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[3]/table/tbody/tr[3]/td[2]/ul/li')
        prcsr_cores.append(procesor_core.text)
    except NoSuchElementException:
        prcsr_cores.append('-----')
    try:
        ram=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        RAM.append(ram.text)
    except NoSuchElementException:
        RAM.append('-----')
    try:
        primary_cam=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        pri_cam.append(primary_cam.text)
    except NoSuchElementException:
        pri_cam.append('-----')
    try:
        second_cam=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        sec_cam.append(second_cam.text)
    except NoSuchElementException:
        sec_cam.append('-----')
    try:
        battery=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[8]/table/tbody/tr/td[2]/ul/li')
        batry.append(battery.text)
    except NoSuchElementException:
        batry.append('-----')
    try:
        prices=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1]')
        price.append(prices.text)
    except NoSuchElementException:
        price.append('--------')
    try:
        rom=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        ROM.append(rom.text)
    except NoSuchElementException:
        ROM.append('--------')
```

```python
for i in range(len(brand)):
    #splitting elements in a brand list based on second appearence of space.
    brand[i]=brand[i].split(" ",2)
    #the 0th element of each row contains brand name,so assigning it to brand list
    brand[i]=brand[i][0]
    
```

```python
flipkart_df=pd.DataFrame()
```

```python
flipkart_df['PRODUCT URL']=url
flipkart_df['brand']=brand
flipkart_df['Smartphone name']=smartphone
flipkart_df['Colour']=color
flipkart_df['RAM']=RAM
flipkart_df['Storage(ROM)']=ROM
flipkart_df['Primary Camera']=pri_cam
flipkart_df['Secondary Camera']=sec_cam
flipkart_df['Display Size']=disp_size
flipkart_df['Display Resolution']=disp_res
flipkart_df['Processor']=prcsr
flipkart_df['processor core']=prcsr_cores
flipkart_df['Battery']=batry
flipkart_df['price']=price
```
