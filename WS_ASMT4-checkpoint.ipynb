```python
#creating driver object
driver = webdriver.Chrome(r"C:/chromedriver.exe")
URL='https://github.com/'
    #getting the driver
driver.get(URL)
```

```python
#getting the url for trending 
TrndPage=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').get_attribute('href')
```

```python
#getting the trending page.
driver.get(TrndPage)
```

```python
urls=[]
```

```python
#adding all the urls of trending profiles to the list named urls.
for i in range(25):
   
   urls.append(driver.find_element_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article['+str(i+1)+']/h1/a').get_attribute('href'))
    
```

```python
urls
```

```python
#creating empty lists to add values later.
title=[]
desc=[]
contributors=[]
languages=[]
```

```python
#looping through each url and extracting the required info.
for i in range(len(urls)):
    driver.get(urls[i])
    try:
        title.append(driver.find_element_by_xpath('//*[@id="js-repo-pjax-container"]/div[1]/div[1]/div/h1/strong/a').text)
    except NoSuchElementException:
        title.append('----')
    try:
        desc.append(driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p').text)
    except NoSuchElementException:
        desc.append('------')
    try:
        a=driver.find_elements_by_xpath('//a[@class="Link--primary no-underline "]')
        #the above xpath reteives multiple info like relases,packages,etc.
        for i in range(len(a)):
            elements=[]
            elements.append(a[i].text)
        #we need contributors,the contributors are usually present at last in the list.
        #After exiting the  for loop,the list named elements holds the last element from xpath which is mostly contributors if contributors are present.
        #In few cases where contibutors are not present,the list may contain null or the last element.
        #print("::",elements)
        
        #the below syntax saves contributors into chk_contribts. 
        chk_contribts = [string for string in elements if 'Contributors' in string]
       
        #if chk_contribts is not  null below stmt executes.
        if len(chk_contribts)!=0:
            contributors.append(chk_contribts)
        
        else:
            #if chk_contribts is null below stmt executes.
            contributors.append('No Contributors')
        
    except NoSuchElementException:
        contributors.append('----')
    try:
        l=[]
        q=driver.find_elements_by_xpath('//span[@class="color-text-primary text-bold mr-1"]')
        for j in range(len(q)):
                l.append(q[j].text)
        languages.append(l)
        
    except NoSuchElementException:
        languages.append('-----')
```

```python
gitTrending_df=pd.DataFrame({'Repository Title':title,
                             'Repository Description':desc,
                             'Contributors count':contributors,
                             'Languages Used':languages
    
})
```

```python
gitTrending_df
```

```python
#closing driver.
driver.close()
```
