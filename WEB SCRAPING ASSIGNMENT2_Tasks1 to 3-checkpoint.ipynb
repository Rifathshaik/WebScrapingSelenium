We need only first 10 jobs info,so storing the text into list.

```python
#converting the required data into text.
#printing the data  as well.
for i in range(10):
    JobInfo[i]=JobInfo[i].text
    print("",JobInfo[i])
    print("*********************************")
```

NOTE:

    2)Company Name
    
    3)Rating if present

    6)Salary
    
    7)Location

```python
JobInfoDF=pd.DataFrame(JobInfo[i].split("\n") for i in range(10))
```

```python
JobInfoDF
```

```python
#Renaming columns.
JobInfoDF.rename(columns={0:'Job Title',1:'COMPANY',2:"RATING",3:"REVIEWS",4:"EXPERIENCE Required",
                         5:"SALARY",6:"LOCATION"},inplace=True)
```

```python
JobInfoDF
```

NOTE:

We can note that for row 2 ,there are no salary and location values.

we can also notice that the values are displaced.

copying experience ,rating and salary values into location,experience and salary fields of 2nd rows.

After copying the values,will be Replacing the entries in Rating and review columns with Not Available.


```python
rows=[1]
for i in range(1):
    j=rows[i]
    JobInfoDF['LOCATION'][j]=JobInfoDF['EXPERIENCE Required'][j]
    JobInfoDF['EXPERIENCE Required'][j]=JobInfoDF['RATING'][j]
    JobInfoDF['SALARY'][j]=JobInfoDF['REVIEWS'][j]
    JobInfoDF['RATING'][j]='Not Available'
    JobInfoDF['REVIEWS'][j]='Not Available'
```

```python
JobInfoDF
```

```python
#Adding job description
JobDescr=driver.find_elements_by_xpath("//div[@class='job-description fs12 grey-text']")
```

```python
#checking how many jobs info we have.
len(JobDescr)
```

```python
JobDescr[0].text
```

```python
#storing the text
for i in range(10):
    JobDescr[i]=JobDescr[i].text
```

```python
#Adding the list as a column to the data frame.

JobInfoDF['Job Description']=JobDescr[0:10]
```

```python
JobInfoDF
```

```python
JobSkills=driver.find_elements_by_xpath("//ul[@class='tags has-description']")
```

```python
len(JobSkills)
```

```python
JobSkills[0].text
```

```python
for i in range(10):
    JobSkills[i]=JobSkills[i].text.replace("\n",",")
```

```python
JobSkills[0]
```

```python
#saving job skills as separate column named "Required Skills" into the data frame.

JobInfoDF['Required Skills']=JobSkills[0:10]
```

```python
JobInfoDF
```

```python
#removing (,) in reviews column.
JobInfoDF['REVIEWS']=JobInfoDF['REVIEWS'].str.replace("(","")
JobInfoDF['REVIEWS']=JobInfoDF['REVIEWS'].str.replace(")","")
```

```python
#Final DATA FRAME
JobInfoDF
```

```python
#closing the driver,After executing the statement,the web page wil be closed.
driver.close()
```
