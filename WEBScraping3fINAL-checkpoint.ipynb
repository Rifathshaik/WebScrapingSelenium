```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

```

```python
#creating driver object
driver = webdriver.Chrome(r"C:/chromedriver.exe")
URL='https://www.amazon.in/'
    #getting the driver
driver.get(URL)
    #promting the user to enter search item
Item=input("Enter The Item you want to search for:")
#passing the search item in search box 
Search = driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']").send_keys(Item)
    #clicking the search button
search_btn = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']").click()
Pagesurl= []
#maximising window as next button was not clicking when window size is small.
driver.maximize_window()
for i in range(3):
         #saves url of present page
        driver.current_url
        Pagesurl.append(driver.current_url)
        next_button = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.PARTIAL_LINK_TEXT, "Next")))
        next_button.click()
        print(i+1,":page URL::",Pagesurl[i])
```

```python
#creating empty listsfor storing urls, Expected delivery, availability, Exchanges
links = []
Expected_Delivery = []
availability = []
ReturnORexchange = []
RatingsNumber = []
Ratings = []
brand = []
product = []
price = []
for i in range(len(Pagesurl)):
    try:
        driver.get(Pagesurl[i])
        print("page no:",i+1)
        products_url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
        print("page no:",i+1,"has:",len(products_url),"products")# finding urls of all items in the present page
        for i in range(len(products_url)):
            links.append(products_url[i].get_attribute("href"))# traversing through each and every products link
    except NoSuchElementException as e:
        continue
    
```

```python
print(len(links))
```

```python
for i in range(len(links)): 
    driver.get(links[i])#getting driver for urls of items stored in links.
    #as brand name and cost are must presentfor every item, not placed them intry block
    productname = driver.find_element_by_xpath("//*[@id='productTitle']")
    product.append(productname.text)# storing the text format of extracted data into list named Bname_1.#In the Bname list, we get titles of each item,if we notice the brand name comes first followed by description of item.
       # Bname[j] = Bname[j].split(" ", 1)# splitting each list values into 2 based on 1 st occuring space.
        #brand.append(Bname[j][0])# Appending the 1 st element of each row to brand
        #product.append(Bname[j][1])## Appending 2 nd element of each row to product.
    try:
        brandname=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[35]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(brandname.text)
    except NoSuchElementException:
        brand.append('---')
    try:
        Cost = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[9]/div[1]/div/table/tbody/tr[2]/td[2]/span[1]')
        price.append(Cost.text)
    except NoSuchElementException as e:
        price.append("---")
    try:
        NoOfRatings = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[5]/div/span[3]/a/span")
        RatingsNumber.append(NoOfRatings.text)
    except NoSuchElementException as e:
        RatingsNumber.append("---")
    try:
        ratings = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[5]/div/span[1]/span/span[1]/a/i[1]/span")
        Ratings.append(ratings.text)
    except NoSuchElementException as e:
        Ratings.append("---")
    try:
        Delivery = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[11]/div/div/b")
        Expected_Delivery.append(Delivery.text)
    except NoSuchElementException as e:
        Expected_Delivery.append("NO DATA")
    try:
        Exchange = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[19]/div[2]/div/div/div/div[2]/span/div[2]/a")
        ReturnORexchange.append(Exchange.text)
    except NoSuchElementException as e:
        ReturnORexchange.append("NO DATA")
    try:
        Availability = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[21]/div[1]/span")
        availability.append(Availability.text)
    except NoSuchElementException as e:
        availability.append("NO DATA")

print("BrandName",len(brand))
print("product name:",len(product))
print("Rating",len(Ratings))
print("ofRatings",len(RatingsNumber))
print("Price",len(price))
print("Return/Excahnge",len(ReturnORexchange))
print("Expected Delivery",len(Expected_Delivery))
print("Availability",len(availability))
print("Product URL",len(links))
   
```
